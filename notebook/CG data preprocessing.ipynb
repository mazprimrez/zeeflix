```python
import pandas as pd
import ast
import numpy as np
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/ðŸ’¡Learning Artificial Intelligence/Tensorflow Recommender/CG_result.csv')
```

```python
rating = pd.read_csv("/content/drive/MyDrive/ðŸ’¡Learning Artificial Intelligence/Tensorflow Recommender/ratings.csv")
```

```python
dataset['movieId'] = dataset['movieId'].apply(ast.literal_eval)
dataset['recs@20'] = dataset['recs@20'].apply(ast.literal_eval)
```

```python
CG_result = dataset.explode('movieId')[['userId', 'movieId']]
CG_result['label'] = 1
```

```python
dummy = dataset.explode('recs@20')[['userId', 'recs@20']]
dummy['label'] = 0
dummy = dummy.rename(columns={'recs@20':'movieId'})
```

```python
df = pd.concat([CG_result, dummy], axis=0)
```

```python
df.drop_duplicates(subset=['userId', 'movieId'], keep='first', inplace=True)
```
