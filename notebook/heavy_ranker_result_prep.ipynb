```python
dictfilt = lambda x, y: dict([ (i,x[i]) for i in x if i in set(y) ])

wanted_keys = ast.literal_eval(dataset['movies'])
wanted_keys = [int(ele) for ele in wanted_keys]
mvs = dictfilt(metadata, wanted_keys)
meta = pd.DataFrame.from_dict(mvs, orient='index')
df = pd.DataFrame({'movieId': [int(ele) for ele in ast.literal_eval(dataset['movies'])], 'rating': ast.literal_eval(dataset['rating'])}).set_index('movieId')
df = df.merge(meta, left_index=True, right_index=True)
```
